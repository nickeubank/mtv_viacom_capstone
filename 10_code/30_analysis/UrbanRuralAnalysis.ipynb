```python
import geopandas as gpd
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data=gpd.read_file("../../20_intermediate_files/30_campuses_w_dist_to_nearest_pp.geojson")
```

```python
data['Region'].unique()
```

```python
### Add Desired Metrics Here

#Percent Cutoff to declare urban/rural
co = 75

#Define urban dataframe
urban = data[data['POPPCT_URBAN'] > co]

#Define rural dataframe
rural = data[data['POPPCT_RURAL'] > co]
```

```python
# Data
df=urban[['distances_2012', 'distances_2016', 'distances_2018', 'distances_2020', 'distances_2020_early', 'Region']]
  
plt.figure(figsize=(12,8))
plt.title('Mean Distance To PP For Urban Schools Over Time in Miles')
plt.plot(df[df['Region'] == 'West'].mean()/1609,label='West')
plt.plot(df[df['Region'] == 'East'].mean()/1609,label = 'East')
plt.plot(df[df['Region'] == 'Mid West'].mean()/1609,label="Mid West")
plt.plot(df[df['Region'] == 'South'].mean()/1609,label="Mid West")
plt.plot(df[df['Region'] == 'North East'].mean()/1609,label="Mid West")

# show legend
plt.legend()

# show graph
plt.show()
```

```python
# Data
df=urban[['distances_2012', 'distances_2016', 'distances_2018', 'distances_2020', 'distances_2020_early', 'Region']]
  
plt.figure(figsize=(12,8))
plt.title('Median Distance To PP For Urban Schools Over Time in Miles')
plt.plot(df[df['Region'] == 'West'].median()/1609,label='West')
plt.plot(df[df['Region'] == 'East'].median()/1609,label = 'East')
plt.plot(df[df['Region'] == 'Mid West'].median()/1609,label="Mid West")
plt.plot(df[df['Region'] == 'South'].median()/1609,label="Mid West")
plt.plot(df[df['Region'] == 'North East'].median()/1609,label="Mid West")

# show legend
plt.legend()

# show graph
plt.show()
```
