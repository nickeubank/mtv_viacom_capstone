```python
import pandas as pd
import numpy as np
```

```python
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
df = pd.read_csv('./mtv_viacom_capstone/20_intermediate_files/30_campuses_w_dist_to_nearest_pp.csv')
df.head()
```

```python
df['Institution Type'].value_counts()
```

```python
#split into public/private and years 
df[['public-private', 'years']] = df['Institution Type'].str.split(",", expand = True)
df[['public-private', 'years']].astype(str)

# set lower case
df['public-private'] = df['public-private'].str.lower()
df['years'] = df['years'].str.lower()
```

```python
# remove white space 
df['public-private'] = df['public-private'].str.replace(' ', '')
df['years'] = df['years'].str.replace(' ', '')
```

```python
df.years.value_counts()
```

```python
# convert distances from meters to miles
df['distances_2020'] = df['distances_2020'] / 1609.34
df['distances_2020_early'] = df['distances_2020_early'] / 1609.34


four_year = df[df['years'] == '4year']
two_year = df[df['years'] == '2year']
```

# 4 Year Schools Stats -- Same Day Voting 

```python
four_year.head()
```

### percent of 4 year schools with polling distance less than 1 mile

```python
under1mile = four_year[four_year['distances_2020'] < 1]
len(under1mile)
```

```python
len(under1mile) / len(four_year) * 100
```

### percent of 4 year schools with polling distance between 1 and 3 miles

```python
between1and3 = four_year[(four_year['distances_2020'] >= 1) & (four_year['distances_2020'] <= 3)]
len(between1and3)
```

```python
len(between1and3) / len(four_year) * 100
```

### percent of 4 year schools with polling distance between 3 and 5 miles

```python
between3and5 = four_year[(four_year['distances_2020'] >= 3) & (four_year['distances_2020'] <= 5)]
len(between3and5)
```

```python
len(between3and5) / len(four_year) * 100
```

### percent of 4 year schools with polling distance over 5 miles

```python
over5miles = four_year[four_year['distances_2020'] > 5]
len(over5miles)
```

```python
len(over5miles) / len(four_year) * 100
```
