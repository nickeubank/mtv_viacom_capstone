```python
s1=sum(data_early.query('distance_2020_early_miles >1')['Total_Enrollment'])/sum(data_early['Total_Enrollment'])
print(str(round(s1*100,2))+" % of students have their nearest early voting option greater than 1 mile from the edge of campus")
```

```python
s2=sum(data_early.query('distance_2020_early_miles >5')['Total_Enrollment'])/sum(data_early['Total_Enrollment'])
print(str(round(s2*100,2))+" % of students have their nearest early voting option greater than 5 mile from the edge of campus")
```

```python
s3=sum(data_early.query('distance_2020_early_miles !=0')['Total_Enrollment'])/sum(data_early['Total_Enrollment'])
print(str(round(s3*100,2))+" % of students do not have an on campus early voting option.")
```

```python
data_early['starbucks_vs_distance_early']=data_early['distances_starbucks']-data_early['distances_2020_early']
s4=sum(data_early[data_early['starbucks_vs_distance_early']<0]['Total_Enrollment'])/sum(data_early['Total_Enrollment'])
print(str(round(s4*100,2))+" % of students have a Starbucks closer than their nearest Early Voting Option.")
```

```python
data_early['target_vs_distance_early']=data_early['distances_target']-data_early['distances_2020_early']
s5=sum(data_early[data_early['target_vs_distance_early']<0]['Total_Enrollment'])/sum(data_early['Total_Enrollment'])
print(str(round(s5*100,2))+" % of students have a Target closer than their nearest Early Voting Option.")
```
