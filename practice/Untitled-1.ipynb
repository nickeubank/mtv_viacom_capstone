```python
import geopandas as gpd
import pandas as pd
```

```python
data=pd.read_csv('../20_intermediate_files/30_campuses_w_dist_to_nearest_pp.csv')
```

```python
data.columns[-100:]
```

```python
data[data['STATE']=='CT'][['Latitude_2020_early', 'Longitude_2020_early','centroid_lat','centroid_long' , '2020_early_distance_by_walking (miles)']]
```

```python
import numpy as np
for i, r in data.iterrows():
    if pd.isna(data.loc[i,'Latitude_2020_early']) or pd.isna(data.loc[i,'Longitude_2020_early']):
        data.loc[i,'2020_early_distance_by_walking (miles)']=np.nan
        data.loc[i,'2020_early_duration_by_walking (minutes)']=np.nan
        data.loc[i,'2020_early_distance_by_driving (miles)']=np.nan
        data.loc[i,'2020_early_duration_by_driving (minutes)']=np.nan
        data.loc[i,'2020_early_distance_by_transit (miles)']=np.nan
        data.loc[i,'2020_early_duration_by_transit (minutes)']=np.nan

    if pd.isna(data.loc[i,'Latitude_2020']) or pd.isna(data.loc[i,'Longitude_2020']):
        data.loc[i,'2020_distance_by_walking (miles)']=np.nan
        data.loc[i,'2020_duration_by_walking (minutes)']=np.nan
        data.loc[i,'2020_distance_by_driving (miles)']=np.nan
        data.loc[i,'2020_duration_by_driving (minutes)']=np.nan
        data.loc[i,'2020_distance_by_transit (miles)']=np.nan
        data.loc[i,'2020_duration_by_transit (minutes)']=np.nan
```

```python
data.groupby('STATE')['2020_early_distance_by_walking (miles)'].mean().to_frame().sort_values(by= '2020_early_distance_by_walking (miles)', ascending = False)
```
